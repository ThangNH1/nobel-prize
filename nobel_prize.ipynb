```python
import pandas as pd
import numpy as np
data = pd.read_csv('./nobel.csv')
data.head()
```

# 1. USA dominance, visualized: to see when the USA started to dominate the Nobel charts we need a plot

```python
data.birth_country.value_counts()
```

```python
data['round_year'] = (np.floor(data['year'] / 10) * 10).astype(int)
data
```

```python
data['usa_winner'] = data["birth_country"] == "United States of America"
data
```

```python
usa_dominance = data.groupby('round_year', as_index=False)['usa_winner'].mean()
usa_dominance
```

```python
import matplotlib.pyplot as plt
plt.xlabel('decade')
plt.ylabel('usa_born_winner')
plt.grid(True)
plt.plot(usa_dominance['round_year'], usa_dominance['usa_winner'])
```

## 2. What is the gender of a typical Nobel Prize winner?

```python
data['female'] = data['sex'] == 'Female'
data
```

```python
female_winner = data.groupby(['category','round_year'], as_index=False)['female'].mean()
female_winner
```

```python
data.category.value_counts()
```

```python
chemistry = female_winner.loc[female_winner['category'] == 'Chemistry']
physics = female_winner.loc[female_winner['category'] == 'Physics']
medicine = female_winner.loc[female_winner['category'] == 'Medicine']
peace = female_winner.loc[female_winner['category'] == 'Peace']
literature = female_winner.loc[female_winner['category'] == 'Literature']
economics = female_winner.loc[female_winner['category'] == 'Economics']
```

```python
plt.xlabel('decade')
plt.ylabel('female_winner')
plt.grid(True)
plt.plot(chemistry['round_year'], chemistry['female'])
plt.plot(physics['round_year'], physics['female'])
plt.plot(medicine['round_year'], medicine['female'])
plt.plot(peace['round_year'], peace['female'])
plt.plot(literature['round_year'], literature['female'])
plt.plot(economics['round_year'], economics['female'])
plt.legend(['Chemistry', 'Physics', 'Medicine', 'Peace', 'Literature', 'Economics'])
```
